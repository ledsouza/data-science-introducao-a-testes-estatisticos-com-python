```python
import pandas as pd
import seaborn as sns
```

```python
tmdb = pd.read_csv('dados/tmdb_5000_movies.csv')
tmdb.head()
```

```python
tmdb.describe()
```

```python
ax = sns.histplot(data = tmdb.vote_average, kde=True, stat='density')
ax.set(xlabel='Nota média', ylabel='Densidade')
ax.set_title('Média de votos em filmes no TMBD 5000')
ax.plot()
```

```python
ax = sns.histplot(data = tmdb.vote_average, kde=True, stat='count')
ax.set(xlabel='Nota média', ylabel='Contagem')
ax.set_title('Média de votos em filmes no TMBD 5000')
ax.plot()
```

```python
ax = sns.boxplot(tmdb.vote_average, orient='h')
ax.set(xlabel='Nota  média do filme')
ax.set_title('Distribuição de nota média dos filmes do TMDB 5000')
ax.plot()
```

A observação de média de votos 0 ou 10 não fazem sentido, será avaliado esses dados:

```python
tmdb.query('vote_average == 0')
```

```python
tmdb.query('vote_average == 10')
```

Observa-se que os filmes que foram avaliados com nota 0 ou 10 possuíam uma contagem de votos nula ou muito pequena. Portanto, optou-se por remover esses dados:

```python
tmdb_com_mais_de_10_votos = tmdb.query('vote_count >= 10')
tmdb_com_mais_de_10_votos.describe()
```

```python
ax = sns.histplot(data = tmdb_com_mais_de_10_votos.vote_average, kde=True, stat='count')
ax.set(xlabel='Nota média', ylabel='Contagem')
ax.set_title('Média de votos em filmes no TMBD 5000')
ax.plot()
```

```python
ax = sns.histplot(data = tmdb_com_mais_de_10_votos.vote_average, kde=True, stat='density')
ax.set(xlabel='Nota média', ylabel='Densidade')
ax.set_title('Média de votos em filmes no TMBD 5000')
ax.plot()
```

```python
ax = sns.histplot(data = tmdb_com_mais_de_10_votos.vote_average, kde=True, stat='density', cumulative=True)
ax.set(xlabel='Nota média', ylabel='% cumulativa de notas')
ax.set_title('Média de votos em filmes no TMBD 5000')
ax.plot()
```

```python
ax = sns.boxplot(tmdb_com_mais_de_10_votos.vote_average, orient='h')
ax.set(xlabel='Nota  média do filme')
ax.set_title('Distribuição de nota média dos filmes do TMDB 5000')
ax.plot()
```

# Análise de MovieLens

```python
notas = pd.read_csv('dados/ratings.csv')
notas.head()
```

```python
nota_media_por_filme = notas.groupby("movieId").mean()["rating"]
nota_media_por_filme.head()
```

```python
ax = sns.histplot(data = nota_media_por_filme.values, kde=True, stat='density')
ax.set(xlabel='Nota média', ylabel='Densidade')
ax.set_title('Média de votos em filmes no MovieLens')
ax.plot()
```

```python
quantidade_de_votos_por_filme = notas.groupby("movieId").count()
quantidade_de_votos_por_filme.query("rating >= 10")
filmes_com_pelo_menos_10_votos = quantidade_de_votos_por_filme.query("rating >= 10").index
```

```python
nota_media_dos_filmes_com_pelo_menos_10_votos = nota_media_por_filme.loc[filmes_com_pelo_menos_10_votos.values]
nota_media_dos_filmes_com_pelo_menos_10_votos.head()
```

```python
ax = sns.histplot(data = nota_media_dos_filmes_com_pelo_menos_10_votos.values, kde=True, stat='density')
ax.set(xlabel='Nota média', ylabel='Densidade')
ax.set_title('Média de votos em filmes no MovieLens')
ax.plot()
```

```python
ax = sns.histplot(data = nota_media_dos_filmes_com_pelo_menos_10_votos.values, kde=True, stat='density', cumulative=True)
ax.set(xlabel='Nota média', ylabel='% acumulada de filmes')
ax.set_title('Média de votos em filmes no MovieLens')
ax.plot()
```

```python
ax = sns.boxplot(nota_media_dos_filmes_com_pelo_menos_10_votos.values, orient='h')
ax.set(xlabel='Nota  média do filme')
ax.set_title('Distribuição de nota média dos filmes do MovieLens')
ax.plot()
```

# Analisando a distribuição dos dados capturados de outros campos do TMDB

```python
ax = sns.histplot(data = tmdb_com_mais_de_10_votos.vote_count, kde=True, stat='density')
ax.set(xlabel='N° de votos', ylabel='Densidade')
ax.set_title('N° de votos em filmes no TMBD 5000')
ax.plot()
```

Espera-se esse comportamento, visto que os filmes mais populares serão vistos por muito mais gente. Consequentemente, terão um número maior de votos. Quanto aos filmes pouco populares, será visto por muito menos gente. Portanto, terão bem menos votos. Visto que existem apenas poucos filmes populares, a distribuição se concentra em um número pequeno de votos.

```python
orcamento = tmdb_com_mais_de_10_votos.query('budget > 0')['budget'] # o critério de query é para remover missings
ax = sns.histplot(data = orcamento, kde=True, stat='density')
ax.set(xlabel='Orçamento', ylabel='Densidade')
ax.set_title('Orçamento em filmes no TMBD 5000')
ax.plot()
```

```python
ax = sns.histplot(data = tmdb['popularity'], kde=True, stat='density')
ax.set(xlabel='Popularidade', ylabel='Densidade')
ax.set_title('Popularidade em filmes no TMBD 5000')
ax.plot()
```

Essa distribuição corrobora o que foi dito anteriormente a respeito de que existem poucos filmes com alta popularidade.

```python
ax = sns.histplot(data = tmdb.query('runtime > 0')['runtime'], kde=True, stat='density')
ax.set(xlabel='Tempo de duração', ylabel='Densidade')
ax.set_title('Tempo de duração em filmes no TMBD 5000')
ax.plot()
```

```python
ax = sns.histplot(data = tmdb.query('runtime > 0')['runtime'], kde=True, stat='density', cumulative=True)
ax.set(xlabel='Tempo de duração', ylabel='% de filmes')
ax.set_title('Tempo de duração em filmes no TMBD 5000')
ax.plot()
```
